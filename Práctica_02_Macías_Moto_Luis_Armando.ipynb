```python
import numpy as np
import matplotlib.pyplot as plt

# Creamos dos arrays de numpy, xn y yn, que representan la temperatura en °F (variable independiente) y el número de chirridos de grillos (variable dependiente)
xn = np.array([64, 50, 52, 51, 57, 55, 60, 63, 63])
yn = np.array([23, 37, 45, 46, 58, 67, 79, 80, 89])

# n almacena la cantidad de puntos de datos
n = len(xn)

# Calculamos las sumas necesarias para las fórmulas de ajuste lineal
Sum_x = sum(xn)
Sum_y = sum(yn)
Sum_xx = sum(xn**2)
Sum_xy = sum(xn*yn)

print(Sum_x, Sum_y, Sum_xx, Sum_xy)

# Calculamos el término independiente de la recta de ajuste lineal (valor de y cuando x es 0)
a_0 = (Sum_xx * Sum_y - Sum_xy * Sum_x) / (n * Sum_xx - Sum_x**2)

# Calculamos la pendiente de la recta de ajuste (cambio en y por unidad de cambio en x)
a_1 = (n * Sum_xy - Sum_x * Sum_y) / (n * Sum_xx - Sum_x**2)

print(a_0, a_1)

# Creamos un array de 100 puntos, distribuidos uniformemente entre los valores mínimo y máximo de xn
x = np.linspace(min(xn), max(xn), 100)

# Calculamos los valores correspondientes de y usando la ecuación de la recta de ajuste
y = a_0 + a_1*x

# Generamos una gráfica de dispersión
plt.scatter(xn, yn, color = 'r')
plt.grid(linestyle = 'dotted')
plt.plot(x, y, color = 'b')

plt.xlabel('Temperatura (F)', fontsize = 16)
plt.ylabel('Chirridos', fontsize = 16)

plt.show()
```

El proyecto seleccionado aborda la relación entre la tasa de abortos y la criminalidad en diversas ciudades de Estados Unidos.

Gracias a la información obtenida es posible proponer dos modelos lineales que representa la relación entre la Tasa de Criminalidad y la Tasa de Abortos en California y en Nueva York.

##Relación entre la Tasa de Abortos y la Tasa de Criminalidad en California

Dada la información obtenida en el periodo de 1990 a 2000, se obtuvo que existe una relación válida entre 10 y 30 abortos por cada 1,000 mujeres, para tasas de criminalidad entre 500 y 1,000 crímenes por cada 100,000 habitantes.

```python
import numpy as np
import matplotlib.pyplot as plt

#Tasa de criminalidad, dada por la varainte (F_CA) y la tasa de abortos (G_CA).

tasa_criminalidad_CA = np.array([900, 850, 800, 780, 750, 700, 680, 650, 600])
tasa_abortos_CA = np.array([30, 28, 27, 25, 23, 22, 20, 18, 15])

n = len(tasa_criminalidad_CA)

# Suma para realizar el ajuste:
Suma_F = sum(tasa_criminalidad_CA)
Suma_G = sum(tasa_abortos_CA)
Suma_FF = sum(tasa_criminalidad_CA**2)
Suma_FG = sum(tasa_criminalidad_CA * tasa_abortos_CA)

# Cálculo de los coeficientes a y b
b_1 = (n * Suma_FG - Suma_F * Suma_G) / (n * Suma_FF - Suma_F**2)
a_1 = (Suma_FF * Suma_G - Suma_FG * Suma_F) / (n * Suma_FF - Suma_F**2)

print(f"Intercepto (b_1): {b_1}")
print(f"Pendiente (a_1): {a_1}")

# Creación de una línea de ajuste
x = np.linspace(min(tasa_abortos_CA), max(tasa_abortos_CA), 100)
y = b_1 + a_1 * x

# Gráfica de dispersión y la línea de ajuste
plt.scatter(tasa_abortos_CA, tasa_criminalidad_CA, color='r')
plt.plot(x, y, color='b')
plt.xlabel('Tasa de Abortos (por cada 1,000 mujeres)')
plt.ylabel('Tasa de Criminalidad (por cada 100,000 habitantes)')
plt.title('Relación entre Tasa de Criminalidad y Tasa de Abortos en California')
plt.grid(linestyle='dotted')
plt.show()

# Calcula la correlación
correlacion = np.corrcoef(tasa_abortos_CA, tasa_criminalidad_CA)[0, 1]
print(f"Coeficiente de Correlación: {correlacion}")
```

##Relación entre la Tasa de Abortos y la Tasa de Criminalidad en Nueva York

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de la tasa de criminalidad (F_NY) y la tasa de abortos (G_NY) en Nueva York
tasa_criminalidad_NY = np.array([1100, 1050, 1000, 950, 900, 850, 800, 750, 700])
tasa_abortos_NY = np.array([35, 34, 32, 30, 28, 26, 24, 22, 20])

# Número de puntos de datos
n = len(tasa_criminalidad_NY)

# Sumas necesarias para el ajuste lineal
Suma_F = sum(tasa_criminalidad_NY)
Suma_G = sum(tasa_abortos_NY)
Suma_FF = sum(tasa_criminalidad_NY**2)
Suma_FG = sum(tasa_criminalidad_NY * tasa_abortos_NY)

# Cálculo de los coeficientes a y b
b_2 = (Suma_FF * Suma_G - Suma_FG * Suma_F) / (n * Suma_FF - Suma_F**2)
a_2 = (n * Suma_FG - Suma_F * Suma_G) / (n * Suma_FF - Suma_F**2)

print(f"Intercepto (b_2): {b_2}")
print(f"Pendiente (a_2): {a_2}")

# Creación de una línea de ajuste
x = np.linspace(min(tasa_criminalidad_NY), max(tasa_criminalidad_NY), 100)
y = b_2 + a_2 * x

# Gráfica de dispersión y la línea de ajuste
plt.scatter(tasa_criminalidad_NY, tasa_abortos_NY, color='r')
plt.plot(x, y, color='b')
plt.xlabel('Tasa de Criminalidad (por cada 100,000 habitantes)')
plt.ylabel('Tasa de Abortos (por cada 1,000 mujeres)')
plt.title('Relación entre Tasa de Criminalidad y Tasa de Abortos en Nueva York')
plt.grid(linestyle='dotted')
plt.show()

# Calcula la correlación
correlacion = np.corrcoef(tasa_abortos_NY, tasa_criminalidad_NY)[0, 1]
print(f"Coeficiente de Correlación: {correlacion}")
```
